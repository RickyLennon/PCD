```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
from docx import Document

# Cargar el archivo de Nelson Mandela
doc2 = Document("20200413_Nelson_Mandela.docx")

# Leer el contenido
for para in doc2.paragraphs:
    print(para.text)

print("\n\n- Nelson Mandela")

# Cargar el archivo de Expropiación Petrolera
doc = Document("Exp_1938.docx")

print("\n\n\n\n\n")

# Leer el contenido
for para in doc.paragraphs:
    print(para.text)

print("\n- Lázaro Cárdenas")
```

```python
num_palabras_por_parrafo1 = [len(paragraph.text.split()) for paragraph in doc2.paragraphs]

if len(num_palabras_por_parrafo1) > 0:
    promedio_palabras1 = sum(num_palabras_por_parrafo1) / len(num_palabras_por_parrafo1)
else:
    promedio_palabras1 = 0  # Para evitar división por cero si no hay párrafos

print(f"El promedio de palabras por párrafo para el discurso de Nelson Mandela es: {promedio_palabras1:.2f}")

num_palabras_por_parrafo2 = [len(paragraph.text.split()) for paragraph in doc.paragraphs]

if len(num_palabras_por_parrafo2) > 0:
    promedio_palabras2 = sum(num_palabras_por_parrafo2) / len(num_palabras_por_parrafo2)
else:
    promedio_palabras2 = 0  # Para evitar división por cero si no hay párrafos

print(f"El promedio de palabras por párrafo para el discurso de Lázaro Cárdenas es: {promedio_palabras2:.2f}")
```

```python
#Función para poder eliminar stop words de nuestro escrito
def tokens(txt, idioma='spanish'):
    texto_limpio = re.sub(r'[^\w\s]', '', txt.lower())  # Eliminar puntuaciones y convertir a minúsculas
    stop_words = set(stopwords.words(idioma))  # Obtener lista de stopwords del idioma seleccionado
    tokens = word_tokenize(texto_limpio)  # Tokenizar el texto
    texto_filtrado = [word for word in tokens if word not in stop_words]  # Eliminar stopwords
    return texto_filtrado

# Extraer texto de cada párrafo
textos1 = [p.text for p in doc.paragraphs if p.text.strip()]  # Filtra líneas vacías
textos2 = [p.text for p in doc2.paragraphs if p.text.strip()]

# Aplicar la función a cada párrafo
map_libro1 = list(map(tokens, textos1))
map_libro2 = list(map(tokens, textos2))
```

```python
map_libro1 = list(map(tokens, [para.text for para in doc.paragraphs]))
map_libro2 = list(map(tokens, [para.text for para in doc2.paragraphs]))
```

```python
#Convertimos las palabras de los párrafos, sin tokens, en una sola lista.
tokens1 = []
for token in map_libro1:
    tokens1.extend(token)

tokens2 = []
for token in map_libro2:
    tokens2.extend(token)

tokens1
```

```python
from nltk import ngrams

frase1 = 'Las compañías ofrecieron, entonces y en respuesta a mis exhortaciones, aumentar los salarios y mejorar ciertas prestaciones'

n = 3 # un trigrama
trigramas1 = ngrams(frase1.split(), n)
for gramas in trigramas1:
    print (gramas)
```

```python
import networkx as nx

lsita_trigramas1 = list(ngrams(frase1.split(), n))

G = nx.DiGraph()

for gram in lsita_trigramas1:
    G.add_edge(gram[0], gram[1])

plt.figure(figsize=(10, 6))
nx.draw(G, with_labels=True, node_size=3000, font_size=14)
plt.title("Grafo de Trigramas - Cárdenas")
plt.show()
```
